```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
da = pd.read_csv("wine_clean.csv", encoding="ISO-8859-1")
da
```

```python
wine = pd.DataFrame(da)
wine
```

```python
#Top ten by average price
wine_by_country = wine.groupby("country").price.mean()
wine_by_country = pd.DataFrame(wine_by_country)
wine_by_country_top_ten = wine_by_country.nlargest(10,'price')
#wine_by_country_top_ten["price"] = wine_by_country_top_ten["price"].map("${:.2f}".format)
wine_by_country_top_ten
```

```python
wine_chart = wine_by_country_top_ten.plot(kind="bar", title="Top Average Prices for Wine")

wine_chart.set_xlabel("Countries")
wine_chart.set_ylabel("Price(Avg)")

plt.show()
plt.tight_layout()
plt.savefig("Visualizations/top_avg_prices_for_wine.png")
```

```python
wine_count = wine["country"].value_counts()
wine_count = pd.DataFrame(wine_count).reset_index()
wine_count = wine_count.rename(columns = {"index" : "Country",
                                          "country" : "Wine Brands"})

wine_count_top_ten = wine_count.nlargest(10, 'Wine Brands')
wine_count_top_ten
```

```python
#Pie Chart showing the distribution of wine brands among the top ten producing countries
by_country = wine_count_top_ten["Country"]
brand_count = wine_count_top_ten["Wine Brands"]

plt.pie(brand_count, labels = by_country, autopct = "%1.1f%%")
plt.title("Distribution of Wine Brands by Top Producing Wine Countries")
plt.show()
plt.savefig("Visualizations/top_producing_wine_countries.png")
```

```python
#By highest average points
wine_points = wine.groupby("country").points.mean()
wine_points = pd.DataFrame(wine_points)
wine_points
```

```python
#Machine Learning section
X = wine["points"].values.reshape(-1, 1)
y = wine["price"].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
predicted = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predicted)

r2 = r2_score(y_test, predicted)
```

```python
print(mse)
print(r2)
```

```python
model.score(X_test, y_test)
```
